```python
import pandas as pd
import json
import requests
import numpy as np
import geopy.distance
```

```python
df = pd.read_csv("Traffic_Volume_Counts__2014-2019_.csv")
df.head()
```

```python
column_names = list(df.columns)
column_names.append("Lat")
column_names.append("Lng")
```

```python
df_out = pd.DataFrame(columns = column_names )
```

```python
df_out = df.copy()
df_out["Lat"]=""
df_out["Lng"]=""
```

```python
main_test = list(df["Roadway Name"])[0:6]
second_test = list(df["From"])[0:6]
```

```python
def getGeocode(intersection):
    result = {}
    apikey = ##removed to prevent overcharging
    url = f'https://maps.googleapis.com/maps/api/geocode/json?address={intersection},+New+York,+NY&key={apikey}'
    r = requests.get(url)
    r.raise_for_status()
    data = r.json()
    if data['results']:
        result = data['results'][0]
    return result
```

```python
intersection_start = []
intersection_end = []

for i, row in df_out.iterrows():
    main_road = row['Roadway Name']
    from_street = row['From']
    to_street = row['To']
    
    intersection_start.append(main_road + " " + from_street)
    intersection_end.append(main_road + " " + to_street)   
```

```python
df_out["Start Intersection"] = intersection_start
df_out["End Intersection"] = intersection_end
```

```python
intersections = intersection_start + intersection_end
len(intersections)
```

```python
unique_intersections = np.unique(intersections)
```

```python
unique_intersections.shape
```

```python
lat = []
lng = []
output_address = []

for address in unique_intersections:
    
    result = getGeocode(address)
    print(result)
    
    if bool(result):
        print(address)
        lat.append(result["geometry"]['location']['lat'])
        lng.append(result["geometry"]['location']['lng'])
        output_address.append(address)

    
```

```python
print(lat, lng)
```

```python
df_output = pd.DataFrame()
df_output["Lat"]=lat
df_output["Lng"]=lng
df_output["Address"]=output_address
df_output.to_excel("df_output.xlsx")
```

```python
temp_start_lat = []
temp_start_lng = []

temp_end_lat = []
temp_end_lng = []

for i, row in df_out.iterrows():
    temp_start_out = df_output.loc[df_output["Address"]==row["Start Intersection"]]
    
    temp_end_out = df_output.loc[df_output["Address"]==row["End Intersection"]]
    if (temp_end_out.empty) or (temp_start_out.empty):
        temp_start_lat.append(np.nan)
        temp_start_lng.append(np.nan)
            
        temp_end_lat.append(np.nan)
        temp_end_lng.append(np.nan)
    else:
        temp_start_lat.append(temp_start_out["Lat"].values[0])
        temp_start_lng.append(temp_start_out["Lng"].values[0])
            
        temp_end_lat.append(temp_end_out["Lat"].values[0])
        temp_end_lng.append(temp_end_out["Lng"].values[0])
    
    
```

```python
len(temp_start_lat)
```

```python
df_out["Lat Start"] = temp_start_lat
df_out["Lng Start"] = temp_start_lng
df_out["Lat End"] = temp_end_lat
df_out["Lng End"] = temp_end_lng
```

```python
df_out.head()
```

```python
df_out.drop(columns=["Lat","Lng"], inplace = True)
```

```python

temp_distance = []

for i, row in df_out.iterrows():
    coord_start = (row["Lat Start"], row["Lng Start"])
    coord_end = (row["Lat End"], row["Lng End"])

    if row.isnull().values.any():         
        distance_between_start_end = np.nan
    else:
        distance_between_start_end = geopy.distance.distance(coord_start, coord_end).miles
    temp_distance.append(distance_between_start_end)
    
```

```python
df_out["distance"] = temp_distance
```

```python
df_out.to_excel("ny flow data.xlsx")
```
