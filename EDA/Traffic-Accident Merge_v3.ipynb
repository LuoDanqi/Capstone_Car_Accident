```python
traffic_data['Segment Address']=traffic_data.apply(lambda x: x['Roadway Name'] + '-' + x['From'] + '-' + x['To'], axis = 1)
```

```python
columns_to_keep = ['ID', 'Segment ID', 'Roadway Name', 'From', 'To', 'Date', 'Start Intersection', 
                   'End Intersection', 'Lat Start', 'Lng Start', 'Lat End', 'Lng End', 'distance', 'Segment Address']

columns_to_average = ['12:00-1:00 AM', '1:00-2:00AM', '2:00-3:00AM', '3:00-4:00AM', '4:00-5:00AM', 
                      '5:00-6:00AM', '6:00-7:00AM', '7:00-8:00AM', '8:00-9:00AM', '9:00-10:00AM', 
                      '10:00-11:00AM', '11:00-12:00PM', '12:00-1:00PM', '1:00-2:00PM', '2:00-3:00PM',
                      '3:00-4:00PM', '4:00-5:00PM', '5:00-6:00PM', '6:00-7:00PM', '7:00-8:00PM', 
                      '8:00-9:00PM', '9:00-10:00PM', '10:00-11:00PM', '11:00-12:00AM']
```

```python
traffic_data_out = traffic_data.groupby(by='Segment Address')[columns_to_average].mean()
```

```python
traffic_data_out.reset_index(inplace = True)
```

```python
traffic_data_out[columns_to_keep] = traffic_data[columns_to_keep]
```

```python
traffic_data_out['Center Lat'] = traffic_data_out[['Lat Start','Lat End']].mean(axis=1)
traffic_data_out['Center Lng'] = traffic_data_out[['Lng Start','Lng End']].mean(axis=1)
```

```python
traffic_data_out.head()
```

```python
rows_with_nan = traffic_data_out.isnull().any(axis =1)
traffic_data_out = traffic_data_out[~rows_with_nan]
```

```python
import datetime

def seperate_month_date(df):
    
    month_list = []
    year_list = []
    weekday_list = [] 
    day_name = []
    
    date = df['Date']

    
    for row in list(date):
        datee = datetime.datetime.strptime(row, "%m/%d/%Y")
        day_name.append(datetime.datetime.strftime(datee, "%A"))

        year_list.append(datee.year)
        month_list.append(datee.month)
        weekday_list.append(datee.weekday())
    
    df['year'] = year_list
    df['month'] = month_list
    df['weekday'] = weekday_list
    df['day name']  = day_name
    
    return df
```

```python
temp_traffic = temp_traffic.melt(id_vars=columns_to_keep, var_name='hour of day')
```

```python
def convert_time_interval(df):
    
    time_day = []
    
    for i, row in df.iterrows():
        am_pm = row['hour of day'].split(':')[2][-2:]
        hour = row['hour of day'].split(':')[0]
        
        if am_pm == 'AM' and hour == '11':
            time_of_day = 23
        elif am_pm == 'PM' and hour == '11':
            time_of_day = 11
        elif am_pm == 'AM' and hour == '12':
            time_of_day = 0
        elif am_pm == 'PM' and hour == '12':
            time_of_day = hour
        elif am_pm == 'PM':
            time_of_day = int(hour) + 12
        else:
            time_of_day = hour
            
        time_day.append(time_of_day)
    
    df["time of day"] = time_day
```

## Positive Sample Joining

```python
accident_data = pd.read_csv('Data/NY_Accidents_June20.csv')
accident_data.head()
```

```python
accident_data['acc_hr_day'].unique()
```

```python
from scipy import spatial
```

```python
def find_closest_node(accident_row, df):
    '''
    accident_row is one row containing accident data with ID, Start_Lat and Start_Lng
    df is the dataframe containing traffic flow with Center Lat and Center Lng
    df schema:
    - Segment Address
    - Center Lat
    - Center Lng
    '''
    
    df_coordinates = df[['Center Lat', 'Center Lng']]
    
    node_tree = spatial.KDTree(df_coordinates)

    distance_to_nearest_node, nearest_node_index = node_tree.query(node_row[['Start_Lat', 'Start_Lng']])
    nearest_node_ID = df.iloc[nearest_node_index][0]

    return nearest_node_ID, distance_to_nearest_node
```

```python
column_name = ['ID','Start_Lat','Start_Lng']

nearest_id = []
nearest_id_distance = []

for row in accident_data.iterrows():
    node_row = row[1].get(column_name)
    flow_id, distance_to_grid_node = find_closest_node(node_row, traffic_data_out)
    nearest_id.append(flow_id)
    nearest_id_distance.append(distance_to_grid_node)
```

```python
accident_data['Traffic Flow ID'] = nearest_id
accident_data['Distance to Traffic Flow']=nearest_id_distance
```

```python
accident_data['Flow ID - hour'] = accident_data.apply(lambda x: x['Traffic Flow ID'] + '-' + str(x['acc_hr_day']) \
                                                      ,axis = 1 )
```

```python
accident_data.head()
```

```python
grouped_traffic['Flow ID - hour'] = grouped_traffic.apply(lambda x: x['Segment Address'] + '-' + str(x['time of day']) ,axis = 1 )
```

```python
traffic_accident_merged = accident_data.merge(grouped_traffic, how = 'left', left_on = 'Flow ID - hour', 
                                  right_on = 'Flow ID - hour')
traffic_accident_merged.head()
```

```python
traffic_accident_merged.to_csv('Data/NY_Accidents.csv')
```
